```python
import googletrans
from googletrans import Translator

def translateDEtoPL(word):
    from_lang = 'de' 
    to_lang = 'pl'
    translator = Translator(service_urls=['translate.google.com'])
    # print(googletrans.LANGUAGES)
    # translator.translate('안녕하세요.')
    # get_sentence = "Speak 'hello' to initiate the Translation !"
    translation = translator.translate(word, src=from_lang, dest=to_lang)
    print(f"{translation.origin}  -->  {translation.text}")
    return translation.text


print(translateDEtoPL("Bitterorangen"))
```

```python
from configparser import ConfigParser
import pandas as pd
import xml.etree.ElementTree as ET
import os
import requests
import base64

parser = ConfigParser()
_ = parser.read('notebook.cfg')
SHOP_API_TOKEN = parser['shop']['SHOP_API_TOKEN']
SHOP_PRODUCTS_URL = parser['shop']['SHOP_PRODUCTS_URL']

tmpFilename = "productList.xml"     

def getHeaders():
    encoded = base64.b64encode((SHOP_API_TOKEN+":").encode())
    token = f"Basic {encoded.decode()}"
    return {"Authorization": token}

def makeGetRequest(url):

    response = requests.get(url, headers = getHeaders())
    # print(response.text)
    return response.text

def makePutRequest(url, payload):
    response = requests.put(url, payload, headers = getHeaders())
    # print(response.text)
    return response.text

def getProductList():
    productListXmlString = makeGetRequest(SHOP_PRODUCTS_URL)
    return ET.fromstring(productListXmlString)[0]

def getProductDetails(productUrl):
    # print(productUrl)
    productDetailsXmlString = makeGetRequest(productUrl)
    # print(productDetailsXmlString)
    # print(ET.fromstring(productDetailsXmlString)[0])
    return ET.ElementTree(ET.fromstring(productDetailsXmlString))
    
def processProduct(productUrl):
    
    productDetails = getProductDetails(productUrl)
    
    if(productToProcess(productDetails)):
        updateProductDetails(productUrl, productDetails)
    else:
        return
    
def productToProcess(productDetails):
    return indeksToProcess(productDetails) and categoryToProcess(productDetails)
   
def indeksToProcess(productDetails):
    indeks = productDetails.find('*').find('reference').text
    # print(indeks is not None and indeks.isnumeric())
    return indeks is not None and indeks.isnumeric()
       
def categoryToProcess(productDetails):
    categoryUrl = productDetails.find('*').find('id_category_default').get("{http://www.w3.org/1999/xlink}href")   
    # print("imp" != getCategoryName(categoryUrl))
    return "imp" != getCategoryName(categoryUrl)

def getCategoryName(categoryUrl):

    categoryXmlString = makeGetRequest(categoryUrl)
    categoryXml = ET.fromstring(categoryXmlString)
    categoryName = categoryXml.find('*').find('name').find('language').text

    # print(ET.fromstring(productDetailsXmlString)[0])
    return categoryName

def clearProductXmlForSaving(productDetails):
    
    productNode = productDetails.find('*')
    
    m_n = productNode.find('manufacturer_name')
    productNode.remove(m_n)
    
    q_n = productNode.find('quantity')
    productNode.remove(q_n)
    
    return productDetails
    
def saveProduct(productUrl, productDetails):
    
    productReadyToSave = clearProductXmlForSaving(productDetails)
    
    xmlToUpdate = ET.tostring(productReadyToSave.getroot(), encoding='utf8', method='xml')
    # print(xmlToUpdate)
    response = makePutRequest(productUrl, xmlToUpdate)
    print(response)
    
def updateProductDetails(productUrl, productDetails):
    productNode = productDetails.find('*')
    print(productNode.find('name').find('language').text)    
    productDetails.find('*').find('name').find('language').text = "nowy name 5"
    print(productNode.find('name').find('language').text)   
    
    saveProduct(productUrl, productDetails)
    
def processProducts():
    for product in getProductList():
        # print(elem)
        productId = product.attrib['id']
        productUrl = product.get("{http://www.w3.org/1999/xlink}href")
        print(f"Id: {productId}, URL: {productUrl}")
        # processProduct(productUrl)
    
    
testProductUrl = SHOP_PRODUCTS_URL + "/1137"
processProduct(testProductUrl)

# processProducts()

```
